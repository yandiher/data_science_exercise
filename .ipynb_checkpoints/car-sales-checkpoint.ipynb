```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
car_sales = pd.read_csv('car-sales.csv') # untuk baca csv file
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(str).str.replace('[\$\,\.]', '', regex=True).astype(int) # untuk ngehapus ($ \ , \ .) yang ada di kolom
```

```python
car_sales
```

```python
car_sales['Price'].plot();
```

```python
car_sales['Make'] = car_sales['Make'].str.lower() # untuk bikin huruf jadi lowercase
```

```python
car_sales
```

```python
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing
```

```python
car_sales_missing['Odometer'] = car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean()) # untuk ngisi kolom yang kosong pake nilai rata-rata
```

```python
car_sales_missing
```

```python
car_sales_missing_data = car_sales_missing.dropna() # untuk ngehapus baris yang ada nilai kosongnya
car_sales_missing_data
```

```python
seat =  pd.Series([5, 5, 5, 5, 5]) # untuk bikin kolom baru
car_sales['Seat'] = seat # untuk nambahin kolom ke dataframe
car_sales
```

```python
car_sales['Seat'].fillna(5, inplace=True) #fillna untuk mengisi cell kosong, inplace untuk menerapkan
car_sales
```

```python
fuel_economy = [7.5, 9.2, 5.0, 9.5, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5] # nambahin kolom pake python
car_sales['Fuel per 100 KM'] = fuel_economy # masukin kolom ke dataframe
car_sales
```

```python
car_sales['Fuel Used (L)'] = car_sales['Odometer (KM)'] / 100 * car_sales['Fuel per 100 KM'] # menambahkan kolom dengan hasil matematika dari kolom lain
car_sales
```

```python
car_sales['Wheel'] = 4
car_sales
```

```python
car_sales = car_sales.drop('Wheel', axis=1) # drop untuk ngehapus, axis 0 adalah baris, axis 1 adalah kolom
car_sales
```

```python
car_sales = car_sales.sample(frac=1) # sample dipakai untuk mengacak urutan sample
car_sales
```
